```
# notebooks/1_eda_and_feature_engineering.ipynb 상단 예시
import pandas as pd
from sqlalchemy import text
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sys
import os

# 프로젝트 루트 경로 설정 (노트북 파일 위치에 따라 조정)
# 예: 노트북이 project_root/notebooks/ 에 있다면
module_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.utils.config_loader import load_config
from src.database.db_utils import get_db_engine

config = load_config('../config/config.yaml') # config 파일 경로 주의
engine = get_db_engine(config['database'])
table_name = config['data']['table_name']

# 특정 파일 데이터 또는 전체 데이터의 일부 로드
# 예: 'bruteforce.csv'에서 온 데이터 중 'BruteForce_Connect_Attempt'로 분류된 것
query_bruteforce_attempts = f"SELECT * FROM {table_name} WHERE attack_type = 'BruteForce_Connect_Attempt' LIMIT 10000;"
df_bf_attempts = pd.read_sql_query(sql=text(query_bruteforce_attempts), con=engine)

query_legit = f"SELECT * FROM {table_name} WHERE attack_type LIKE 'Legitimate%' LIMIT 10000;"
df_legit = pd.read_sql_query(sql=text(query_legit), con=engine)

print("BruteForce Attempts Sample:")
print(df_bf_attempts.head())
print("\nLegitimate Traffic Sample:")
print(df_legit.head())
```
